```python
import fitz  # PyMuPDF
import re

def search_string_in_pdf(pdf_path, search_string):
    page_numbers = []
    doc = fitz.open(pdf_path)
    for page_num in range(doc.page_count):
        page = doc[page_num]
        text = page.get_text("text")
        if re.search(search_string, text, re.IGNORECASE):
        #if search_string in text:
            page_numbers.append(page_num + 1)  # Page numbers start from 1
    doc.close()
    return page_numbers
```

```python
import os
import pandas as pd

import os
import pandas as pd

def read_txt_files(folder_path):
    data = []
    for filename in sorted(os.listdir(folder_path)):
        if filename.endswith(".txt"):
            file_path = os.path.join(folder_path, filename)
            with open(file_path, "r") as file:
                content = file.read()
                data.append({'File Name': filename, 'Content': content})
    df2 = pd.DataFrame(data)
    return df2

# Example usage:
folder_path = "C:\\Users\\Austraxpc60\\Downloads\\roofing-product-mapping\\All pdf trimmed"
df2 = read_txt_files(folder_path)
df2.to_csv("output.csv", index=False)


```
