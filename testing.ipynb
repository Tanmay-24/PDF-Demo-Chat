```python
import difflib

def compare_documents(doc1, doc2):
    # Split documents into lines
    lines1 = doc1.splitlines()
    lines2 = doc2.splitlines()
    
    # Calculate similarity ratio using SequenceMatcher
    matcher = difflib.SequenceMatcher(None, lines1, lines2)
    similarity_ratio = matcher.ratio()
    
    return similarity_ratio

if __name__ == "__main__":
    # Sample documents containing keywords
    doc1 = """
    SBS Roofing SystemStructurally Flat Metal DeckPrefinished Steel Roof Edge FlashingCoping CapGutters and DownspoutsStructurally Flat Concrete DeckTwo-Part Urethane Insulation Adhesive
    """
    
    doc2 = """SBS Roof SystemModified Bituminous MembraneExisting SBS Membrane Roof OverlayValero Type IV AsphaltTorch ApplicationTapered Polyiso InsulationPre-finished Steel FasciaGravel StopDrip EdgeRoof Walkway PadsBox GutterDownspoutsConcrete Splash BlockRoof Drain Sumps
    """
    
    # Compare documents
    similarity_ratio = compare_documents(doc1, doc2)
    print("Similarity Ratio:", similarity_ratio)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def calculate_similarity(doc1, doc2, keywords):
    # Combine both documents into a list
    documents = [doc1, doc2]

    # Initialize TF-IDF Vectorizer
    vectorizer = TfidfVectorizer(vocabulary=keywords)

    # Fit and transform the documents
    tfidf_matrix = vectorizer.fit_transform(documents)

    # Calculate cosine similarity between the two documents
    similarity = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])[0][0]

    return similarity

if __name__ == "__main__":
    # Sample documents containing keywords
    doc1 = """
    SBS Roofing SystemStructurally Flat Metal DeckPrefinished Steel Roof Edge FlashingCoping CapGutters and DownspoutsStructurally Flat Concrete DeckTwo-Part Urethane Insulation Adhesive
    """
    
    doc2 = """
    24 ga. pre-finished Steelcounterflashingroof edge flashingcoping capflashingtrimguttersdownspouts
    """

    # List of keywords
    keywords = ["Metal", "Roof", "Panels", "Standing", "Seam", "Underlayment", "24", "ga.", "pre-finished", "steel",
                "roof", "related", "flashing", "and", "trim", "3”x3”", "square", "gutter", "2”x2”", "downspout"]

    # Calculate similarity
    similarity = calculate_similarity(doc1, doc2, keywords)
    print("Similarity Score:", similarity)
```
